# Decision Tree Tutorial

```python
import pandas as pd
from sklearn import tree
```

```python
# Read CSV file and save data frame in variable df
df = pd.read_csv("./data.csv", sep=",")

X = df[["Preis"]]     # features
Y = df[["Produkt"]]   # targets
```

```python
# Show data frame
df
```

```python
# Create a new instance of a decision tree
clf = tree.DecisionTreeClassifier()

# Present the decision tree our data. The decision tree will now create a new rule for predicing products from prices
clf = clf.fit(X, Y)
```

```python
tree.plot_tree(clf)
```

### Make Predictions

```python
# Load CSV file 'data_test.csv'
df_test = pd.read_csv("data_test.csv", sep=",")

# Extract column 'Preis'
X_test = df_test[["Preis"]]
```

```python
X_test
```

```python
y_pred = clf.predict(X_test)
```
